```python
# importing libraries

import pandas as pd
import numpy as np
# for pre-processing
from sklearn import preprocessing
from sklearn.model_selection import KFold,cross_val_score

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/train.csv')
```

```python
df = df.dropna()
```

```python
df = df.drop(['ID', 'SSN', 'Name', 'Customer_ID', 'Type_of_Loan', 'Occupation', 'Month', 'Monthly_Balance', 'Amount_invested_monthly'],axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
# use label encoder to convert required columns into numbers

label_encoder = preprocessing.LabelEncoder()
df['Age'] = label_encoder.fit_transform(df['Age'])
df['Annual_Income'] = label_encoder.fit_transform(df['Annual_Income'])
df['Monthly_Inhand_Salary'] = label_encoder.fit_transform(df['Monthly_Inhand_Salary'])
df['Num_of_Loan'] = label_encoder.fit_transform(df['Num_of_Loan'])
df['Num_of_Delayed_Payment'] = label_encoder.fit_transform(df['Num_of_Delayed_Payment'])
df['Changed_Credit_Limit'] = label_encoder.fit_transform(df['Changed_Credit_Limit'])
df['Credit_Mix'] = label_encoder.fit_transform(df['Credit_Mix'])
df['Outstanding_Debt'] = label_encoder.fit_transform(df['Outstanding_Debt'])
df['Credit_History_Age'] = label_encoder.fit_transform(df['Credit_History_Age'])
df['Payment_of_Min_Amount'] = label_encoder.fit_transform(df['Payment_of_Min_Amount'])
df['Payment_Behaviour'] = label_encoder.fit_transform(df['Payment_Behaviour'])
####df['Monthly_Balance'] = label_encoder.fit_transform(df['Monthly_Balance'])
####df['Amount_invested_monthly'] = label_encoder.fit_transform(df['Amount_invested_monthly'])
df['Credit_Score'] = label_encoder.fit_transform(df['Credit_Score'])
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(x.shape)
```

```python
print(y_train.shape)
```

```python
sel = SelectFromModel(RandomForestClassifier(n_estimators = 100,max_features = 3))
sel.fit(x_train,y_train)
```

```python
sel.get_support()
```

```python
selected_feat= x_train.columns[(sel.get_support())]
len(selected_feat)
```

```python
# Creating a Random Forest Classifier model
random_forest_model = RandomForestClassifier(random_state=10)

# Fiting the model to the training data
random_forest_model.fit(x_train, y_train)

# Making predictions on the test data
pred= random_forest_model.predict(x_test)
```

```python
np.mean(y_test==pred)
```

```python
# Calculating accuracy of testing data
accuracy = accuracy_score(y_test, pred)
print("Random Forest Accuracy:", accuracy)
```

```python
# Making predictions on the training data
pred1= random_forest_model.predict(x_train)
```

```python
# Calculating accuracy of training data
accuracy = accuracy_score(y_train, pred1)
print("Random Forest Accuracy:", accuracy)
```
